```
import pandas as pd
import matplotlib.pyplot as plt

df_TimeAge = pd.read_csv("/gdrive/My Drive/한양대학교/2020년 문화관광 빅데이터 분석대회/data/코로나/Data Science for COVID-19 (DS4C)/TimeAge.csv", encoding = 'cp949')

print(df_TimeAge)

df_TimeAge_0 = df_TimeAge[df_TimeAge["age"]=="0s"]
df_TimeAge_10 = df_TimeAge[df_TimeAge["age"]=="10s"]
df_TimeAge_20 = df_TimeAge[df_TimeAge["age"]=="20s"]
df_TimeAge_30 = df_TimeAge[df_TimeAge["age"]=="30s"]
df_TimeAge_40 = df_TimeAge[df_TimeAge["age"]=="40s"]
df_TimeAge_50 = df_TimeAge[df_TimeAge["age"]=="50s"]
df_TimeAge_60 = df_TimeAge[df_TimeAge["age"]=="60s"]
df_TimeAge_70 = df_TimeAge[df_TimeAge["age"]=="70s"]
df_TimeAge_80 = df_TimeAge[df_TimeAge["age"]=="80s"]

plt.plot(df_TimeAge_0["confirmed"])
plt.plot(df_TimeAge_10["confirmed"])
plt.plot(df_TimeAge_20["confirmed"])
plt.plot(df_TimeAge_30["confirmed"])
plt.plot(df_TimeAge_40["confirmed"])
plt.plot(df_TimeAge_50["confirmed"])
plt.plot(df_TimeAge_60["confirmed"])
plt.plot(df_TimeAge_70["confirmed"])
plt.plot(df_TimeAge_80["confirmed"])
plt.legend(["0s","10s","20s","30s","40s","50s","60s","70s","80s"])
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt

df_TimeProvince = pd.read_csv("/gdrive/My Drive/한양대학교/2020년 문화관광 빅데이터 분석대회/data/코로나/Data Science for COVID-19 (DS4C)/TimeProvince.csv", encoding = 'cp949')

print(df_TimeProvince)
print(list(set(df_TimeProvince["province"])))

df_TimeProvince_Daegu = df_TimeProvince[df_TimeProvince["province"]=="Daegu"]
df_TimeProvince_Jeollanam_do = df_TimeProvince[df_TimeProvince["province"]=="Jeollanam-do"]
df_TimeProvince_Gwangju = df_TimeProvince[df_TimeProvince["province"]=="Gwangju"]
df_TimeProvince_Seoul = df_TimeProvince[df_TimeProvince["province"]=="Seoul"]
df_TimeProvince_Jeollabuk_do = df_TimeProvince[df_TimeProvince["province"]=="Jeollabuk-do"]
df_TimeProvince_Incheon = df_TimeProvince[df_TimeProvince["province"]=="Incheon"]
df_TimeProvince_Busan = df_TimeProvince[df_TimeProvince["province"]=="Busan"]
df_TimeProvince_Sejong = df_TimeProvince[df_TimeProvince["province"]=="Sejong"]
df_TimeProvince_Gyeonggi_do = df_TimeProvince[df_TimeProvince["province"]=="Gyeonggi-do"]
df_TimeProvince_Ulsan = df_TimeProvince[df_TimeProvince["province"]=="Ulsan"]
df_TimeProvince_Chungcheongbuk_do = df_TimeProvince[df_TimeProvince["province"]=="Chungcheongbuk-do"]
df_TimeProvince_Jeju_do = df_TimeProvince[df_TimeProvince["province"]=="Jeju-do"]
df_TimeProvince_Gyeongsangbuk_do = df_TimeProvince[df_TimeProvince["province"]=="Gyeongsangbuk-do"]
df_TimeProvince_Daejeon = df_TimeProvince[df_TimeProvince["province"]=="Daejeon"]
df_TimeProvince_Chungcheongnam_do = df_TimeProvince[df_TimeProvince["province"]=="Chungcheongnam-do"]
df_TimeProvince_Gyeongsangnam_do = df_TimeProvince[df_TimeProvince["province"]=="Gyeongsangnam-do"]
df_TimeProvince_Gangwon_do = df_TimeProvince[df_TimeProvince["province"]=="Gangwon-do"]

plt.plot(df_TimeProvince_Daegu["confirmed"])
plt.plot(df_TimeProvince_Jeollanam_do["confirmed"])
plt.plot(df_TimeProvince_Gwangju["confirmed"])
plt.plot(df_TimeProvince_Seoul["confirmed"])
plt.plot(df_TimeProvince_Jeollabuk_do["confirmed"])
plt.plot(df_TimeProvince_Incheon["confirmed"])
plt.plot(df_TimeProvince_Busan["confirmed"])
plt.plot(df_TimeProvince_Sejong["confirmed"])
plt.plot(df_TimeProvince_Gyeonggi_do["confirmed"])
plt.plot(df_TimeProvince_Ulsan["confirmed"])
plt.plot(df_TimeProvince_Chungcheongbuk_do["confirmed"])
plt.plot(df_TimeProvince_Jeju_do["confirmed"])
plt.plot(df_TimeProvince_Gyeongsangbuk_do["confirmed"])
plt.plot(df_TimeProvince_Daejeon["confirmed"])
plt.plot(df_TimeProvince_Chungcheongnam_do["confirmed"])
plt.plot(df_TimeProvince_Gyeongsangnam_do["confirmed"])
plt.plot(df_TimeProvince_Gangwon_do["confirmed"])
plt.legend(['Daegu', 'Jeollanam-do', 'Gwangju', 'Seoul', 'Jeollabuk-do', 'Incheon', 'Busan', 'Sejong', 'Gyeonggi-do', 'Ulsan', 'Chungcheongbuk-do', 'Jeju-do', 'Gyeongsangbuk-do', 'Daejeon', 'Chungcheongnam-do', 'Gyeongsangnam-do', 'Gangwon-do'])
plt.show()
```
